```
all_td_np = np.array([])
all_eto_np = np.array([])
all_tc_np = np.array([])
all_ta_np = np.array([])
all_i_np = np.array([])
all_vpd_np = np.array([])
all_avg_td_np = np.array([])
all_avg_eto_np = np.array([])
all_avg_tc_np = np.array([])
all_avg_ta_np = np.array([])
all_avg_i_np = np.array([])
all_avg_vpd_np = np.array([])

#start_date=dt.datetime.now() + dt.timedelta(days=-1)        
start_date=dt.datetime.strptime('2021/01/01','%Y/%m/%d')
#end_date=start_date
end_date=dt.datetime.strptime('2021/01/30','%Y/%m/%d')

date = start_date
while date<=end_date:

  sampling_period_in_sec = 60*5

  time_s,tc_s=GetDailyData('DURIAN-01','16416','100',date.strftime("%Y-%m-%d"))  
  tc_s=ReSamplingDailyData(time_s, tc_s,sampling_period_in_sec)
  
  time_s,ta_s=GetDailyData('TEST-DURIAN-01-WT','4096','1',date.strftime("%Y-%m-%d"))
  ta_s=ReSamplingDailyData(time_s, ta_s,sampling_period_in_sec)

  time_s,i_s = GetDailyData('TEST-DURIAN-01-WT','4096','7',date.strftime("%Y-%m-%d"))
  i_s = ReSamplingDailyData(time_s, i_s,sampling_period_in_sec)  

  time_s,rh_s = GetDailyData('TEST-DURIAN-01-WT','4096','2',date.strftime("%Y-%m-%d"))
  rh_s= ReSamplingDailyData(time_s, rh_s,sampling_period_in_sec)

  time_s,p_s = GetDailyData('TEST-DURIAN-01-WT','4096','8',date.strftime("%Y-%m-%d"))
  p_s = ReSamplingDailyData(time_s, p_s,sampling_period_in_sec)

  time_s,w_s = GetDailyData('TEST-DURIAN-01-WT','4096','5',date.strftime("%Y-%m-%d"))
  w_s = ReSamplingDailyData(time_s, w_s,sampling_period_in_sec)

  time2_s,sm_s = GetDailyData('TEST-DURIAN-01-WT',F'{0x1300}','1100',date.strftime("%Y-%m-%d"))
  sm_s = ReSamplingDailyData(time_s, w_s,sampling_period_in_sec)
  
 
  moving_avg_point = 9
  tc_np = m_avg(np.array(tc_s),moving_avg_point)
  ta_np = m_avg(np.array(ta_s),moving_avg_point)
  i_np =  m_avg(np.array(i_s),moving_avg_point)
  rh_np = m_avg(np.array(rh_s),moving_avg_point)
  p_np = m_avg(np.array(p_s),moving_avg_point)
  w_np = m_avg(np.array(w_s),moving_avg_point)
  sm_np = np.array(sm_s)

  td_np = ta_np-tc_np
  vpd_np = CalcVPD(ta_np,rh_np)
  eto_np = CalcETo(ta_np,rh_np,p_np,i_np,w_np)

  print(F'{date} ETo {np.sum(eto_np):.1f}')

  i_start=int(12.0*12)
  i_stop =int(13.0*12)

  all_eto_np = np.append(all_eto_np,eto_np[i_start:i_stop])
  all_avg_eto_np = np.append(all_avg_eto_np,np.average(eto_np[i_start:i_stop]))

  all_td_np = np.append(all_td_np,td_np[i_start:i_stop])
  all_avg_td_np = np.append(all_avg_td_np,np.average(td_np[i_start:i_stop]))

  all_vpd_np = np.append(all_vpd_np,vpd_np[i_start:i_stop])
  all_avg_vpd_np = np.append(all_avg_vpd_np,np.average(vpd_np[i_start:i_stop]))

  time_np = np.arange(0,24,24/288)

  plt.rcParams['figure.figsize'] = [16, 3]
  fig, axs = plt.subplots(1,4)
  t=-1
  for i in range(0,4) :
    axs[i].title.set_text(F"Tair-Tc vs ETo @t{t*5:+}min")
    axs[i].set(xlim=(0,0.1),ylim=(-3,5))
    axs[i].scatter(eto_np[i_start:i_stop],td_np[i_start+t:i_stop+t])
    t=t+1

  plt.rcParams['figure.figsize'] = [16, 7]
  fig, axs = plt.subplots(2,4)
  
  axs[0,0].title.set_text('ETo all day (mm)')  
  axs[0,0].set(ylim=(0,0.1),xlim=(0,24))
  axs[0,0].plot(time_np,eto_np)

  axs[0,1].title.set_text('Light Intensity all day (w/m2')
  axs[0,1].set(ylim=(0,1200),xlim=(0,24))
  axs[0,1].plot(time_np,i_s)

  axs[0,2].title.set_text('ETo (mm)')
  axs[0,2].set(ylim=(0,0.1))
  axs[0,2].plot(time_np[i_start:i_stop],eto_np[i_start:i_stop])

  axs[0,3].title.set_text('Tair-Tc')
  axs[0,3].set(ylim=(-3,5))
  axs[0,3].plot(time_np[i_start:i_stop],td_np[i_start:i_stop])

  axs[1,0].title.set_text('Wind Velocity (m/s)')
  axs[1,0].set(ylim=(0,4))
  axs[1,0].plot(time_np[i_start:i_stop],w_np[i_start:i_stop])
 
  axs[1,1].title.set_text('Ambient Temp (c)')
  axs[1,1].set(ylim=(20,40))
  axs[1,1].plot(time_np[i_start:i_stop],ta_np[i_start:i_stop])

  axs[1,2].title.set_text('Canopy Temp (c)')
  axs[1,2].set(ylim=(20,40))
  axs[1,2].plot(time_np[i_start:i_stop],tc_np[i_start:i_stop])

  axs[1,3].title.set_text('VPD (kPa)')
  axs[1,3].set(ylim=(1,3))
  axs[1,3].plot(time_np[i_start:i_stop],vpd_np[i_start:i_stop])

  plt.show()
  date=date+dt.timedelta(days=1)

plt.rcParams['figure.figsize'] = [8, 8]
plt.title("Tair-Tc vs ETo")
plt.scatter(all_eto_np,all_td_np)
plt.scatter(all_avg_eto_np,all_avg_td_np)
plt.show()
plt.title("Tair-Tc vs VPD")
plt.scatter(all_vpd_np,all_td_np)
plt.scatter(all_avg_vpd_np,all_avg_td_np)


 
#if __name__ == "__main__":
#  main()
```
